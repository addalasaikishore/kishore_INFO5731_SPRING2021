```python
# You code here 
import requests
import re
import bs4
from bs4 import BeautifulSoup
import pandas as pd


```

step 1:get the website by using request.get method

```python

flip=requests.get("https://www.flipkart.com/search?q=samsung+mob&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_15_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_15_na_na_na&as-pos=1&as-type=RECENT&suggestionId=samsung+mob%7CMobiles&requestId=bb46d944-96bb-4522-8a5f-ea18c8bc85e0&as-searchtext=samsung%20moblies")
```

```python
flip
```

### getting raw data

```python
flip=flip.text
flip
```

### step 3: by using BeautifulSoup converting raw data into structured data

```python
flipkart=BeautifulSoup(flip)
```

```python
flipkart
```

```python
samsung=flipkart.find_all("div",attrs={"class":"_4rR01T"})
```

```python
mobile=[]
for i in samsung:
    mobile.append(i.text)
print(mobile)
```

extracting all samsung moblies and price of moblie

```python
moblies=[]
prices=[]
for i in flipkart.find_all("a",attrs={"class":"_1fQZEK"}):
    for i in flipkart.find_all("div",attrs={"class":"_4rR01T"}):
        moblies.append(i.text)
        for i in flipkart.find_all("div",attrs={"class":"_30jeq3 _1_WHN1"}):
            prices.append(i.text)
print(moblies)
```

```python
prices=prices[0:576]
len(prices)
```

**creating dataframe by using pandas**

```python
data_moblie=pd.DataFrame()
data_moblie["mobile_name"]=moblies
data_moblie["prices"]=prices
```
